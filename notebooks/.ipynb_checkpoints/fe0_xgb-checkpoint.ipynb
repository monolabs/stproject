```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
%matplotlib inline
import seaborn as sns

from hyperopt import hp

import sys
sys.path.insert(1, '../src/stproject')
from utils import *

import pickle
```

```python
df_stliq = pd.read_csv('../data/df_stliq_clean.csv', index_col=0)
df_fe0 = pd.read_csv('../data/df_fe0.csv', index_col=0)
df_fe0['mv'] = df_fe0['M'] / df_fe0['density']

print(f"df_stliq has length: {len(df_stliq)}")
display(df_stliq.head())
print(f"df_raw has length: {len(df_fe0)}")
display(df_fe0.head())
```

```python
# Exclude features globally
features_excluded = ['C#C', 'R3', 'R4', 'R7', 'R8']
df_fe0_clean = df_fe0.loc[df_fe0[features_excluded].sum(axis=1) == 0, df_fe0.columns.difference(features_excluded)]

X_fe0 = df_fe0_clean[df_fe0_clean.columns.difference(['molecule', 'measured_st'])]
y_fe0 = df_fe0_clean['measured_st']

print(f"X_fe0 has length: {len(X_fe0)}")
display(X_fe0.head())
```

```python
# sns.pairplot(df_fe0_clean[df_fe0_clean.columns.difference(['molecule'])])
```

```python
# Splitting out test set
# Test set is chosen according to literature (original_id in df_stliq contains 'Test')
test_idx = df_stliq[df_stliq['original_id'].str.contains('Test')].index.tolist()
X_fe0_train = X_fe0.loc[~X_fe0.index.isin(test_idx)]
X_fe0_test = X_fe0.loc[test_idx]
y_fe0_train = y_fe0.loc[~X_fe0.index.isin(test_idx)]
y_fe0_test = y_fe0.loc[test_idx]

print(f"X_fe0_train has length: {len(X_fe0_train)}")
print(f"X_fe0_test has length: {len(X_fe0_test)}")
display(X_fe0_test.head())
```

```python
# defining space of optimization with BO
space = {
             'n_estimators' : hp.quniform('n_estimators', 100, 1000, 1),
             'eta' : hp.quniform('eta', 0.01, 0.10, 0.02),
             'max_depth' : hp.choice('max_depth', np.arange(1, 13, 1)),
             'min_child_weight' : hp.quniform('min_child_weight', 1, 6, 1),
             'subsample' : hp.quniform('subsample', 0.5, 1, 0.05),
             'gamma' : hp.quniform('gamma', 0.5, 1, 0.05),
             'colsample_bytree' : hp.quniform('colsample_bytree', 0.5, 1, 0.05),
             'eval_metric': 'rmse',
             'objective': 'reg:squarederror',
             'nthread' : 6,
             'silent' : 1,
             'seed': 42
             }
```

```python
rmse_fe0_xgb_f_ll, mae_fe0_xgb_f_ll, i_out_fe0_xgb_f_ll = evaluate_xgb_model(model_fe0_xgb_f_l,
                        X_fe0_test, 
                        y_fe0_test,
                        num_outliers=5, 
                        title='fe0_xgb_f_ll\nfull features, train: liquid, test: liquid',
                        x_label='predicted surface tension',
                        y_label='measured surface tension',
                        with_line=True)
print(f"rmse = {rmse_fe0_xgb_f_ll}, mae = {mae_fe0_xgb_f_ll}")
display(df_fe0.loc[i_out_fe0_xgb_f_ll])
```

```python
# plotting feature importances
fi_fe0_xgb_f_l = pd.DataFrame(model_fe0_xgb_f_l.get_score(importance_type='gain'), index=[0])
display(fi_fe0_xgb_f_l)
sns.barplot(x=fi_fe0_xgb_f_l.loc[0], y=fi_fe0_xgb_f_l.columns)
```

### fe0_xgb_mv_ll

XGB with mv-scaled fragments, trained on liquids, tested on liquids

```python
X_fe0_train_mvscaled = (X_fe0_train[X_fe0_train.columns.difference(['M', 'mv', 'density'])]
                       .divide(X_fe0_train['mv'], axis=0))

X_fe0_test_mvscaled = (X_fe0_test[X_fe0_test.columns.difference(['M', 'mv', 'density'])]
                       .divide(X_fe0_test['mv'], axis=0))

model_fe0_xgb_mv_l, params_fe0_xgb_mv_l, trials_fe0_xgb_mv_l = run_xgb(
    X_fe0_train_mvscaled,
    y_fe0_train,
    space,
    k=5,
    max_evals=250,
    random_state=42)
```

```python
with open('../models/model_fe0_xgb_mv_l.pickle', 'wb') as handle:
    pickle.dump(model_fe0_xgb_mv_l, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('../models/params_fe0_xgb_mv_l.pickle', 'wb') as handle:
    pickle.dump(params_fe0_xgb_mv_l, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('../models/trials_fe0_xgb_mv_l.pickle', 'wb') as handle:
    pickle.dump(trials_fe0_xgb_mv_l, handle, protocol=pickle.HIGHEST_PROTOCOL)

# post-training (pickle of models exists)
with open('../models/model_fe0_xgb_mv_l.pickle', 'rb') as handle:
    model_fe0_xgb_mv_l = pickle.load(handle)
```

```python
rmse_fe0_xgb_mv_ll, mae_fe0_xgb_mv_ll, i_out_fe0_xgb_mv_ll = evaluate_xgb_model(model_fe0_xgb_mv_l,
                        X_fe0_test_mvscaled, 
                        y_fe0_test,
                        num_outliers=5, 
                        title='fe0_xgb_mv_ll\nmv-scaled, train: liquid, test: liquid',
                        x_label='predicted surface tension',
                        y_label='measured surface tension',
                        with_line=True)
print(f"rmse = {rmse_fe0_xgb_mv_ll}, mae = {mae_fe0_xgb_mv_ll}")
display(df_fe0.loc[i_out_fe0_xgb_mv_ll])
```
