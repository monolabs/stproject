```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics

import sys
sys.path.insert(1, '../src/stproject')
from utils import *

import pickle
```

```python
# importing feature-engineered dataset v1 (df_fe1.csv)

df_fe0 = pd.read_csv('../data/df_fe0.csv', index_col=0)
frags = df_fe0.columns.difference(['M', 'molecule', 'measured_st', 'density']).tolist()

# computing new feature: molar volume (mv = M/density)
df_fe0['mv'] = df_fe0['M']/df_fe0['density']
print(len(df_fe0))
frags
```

## fe0 lm0 - molar volume (train: liquid, pred: liquid)

```python
lm0, rmse0, mae0, i_out0 = run_linear_regression(df_fe0, 
                                     set(frags)-set(['C#C', 'R3', 'R4', 'R7', 'R8']), 
                                     ['C#C', 'R3', 'R4', 'R7', 'R8'], 
                                     'mv',
                                     fit_intercept=True,
                                     num_outliers=5,
                                     title='fe0 lm0 - molar volume\n(train: liquid, pred: liquid)',
                                     x_label='predicted molar volume',
                                     y_label='molar volume calculated from density',
                                     with_line=True)
print(f"rmse0 = {rmse0}, mae0 = {mae0}")
df_fe0.loc[i_out0]
```

## fe0 lm0s - molar volume (train: liquid, pred: liquid, short)

```python
lm0s, rmse0s, mae0s, i_out0s = run_linear_regression(df_fe0, 
                      set(frags)-set(['C#C', 'R3', 'R4', 'R7', 'R8', 'O-alc', 'O-ald', 'O-acid']), 
                      ['C#C', 'R3', 'R4', 'R7', 'R8', 'O-alc', 'O-ald', 'O-acid'], 
                      'mv',
                      fit_intercept=True,
                      num_outliers=5,
                      title='fe0 lm0s - molar volume\n(train: liquid, pred: liquid, short)',
                      x_label='predicted molar volume',
                      y_label='molar volume calculated from density',
                      with_line=True)
print(f"rmse0s = {rmse0s}, mae0s = {mae0s}")
df_fe0.loc[i_out0s]
```

## fe0 lm1 - surface tension parachor (train: liquid, pred: liquid)

In this model, surface tension (gamma) is expressed as gamma = (P/mv)^n, with <br> 
mv = M/density <br>
n = power (value of 4 is suggested by literatures)

```python
lm1, rmse1, mae1, i_out1 = run_linear_regression_parachor(df_fe0,
                        set(frags)-set(['C#C', 'R3', 'R4', 'R7', 'R8']), 
                        ['C#C', 'R3', 'R4', 'R7', 'R8'],
                        'measured_st',
                        'mv',
                        n=4,
                        run_on_st=True,
                        fit_intercept=True,
                        num_outliers=5,
                        title='fe0 lm1 - surface tension parachor\n(train: liquid, pred: liquid)',
                        x_label='predicted surface tension',
                        y_label='measured surface tension',
                        with_line=True)
print(f"rmse1 = {rmse1}, mae1 = {mae1}")
df_fe0.loc[i_out1]
```

## fe0 lm1x - surface tension parachor (train: liquid, pred: liquid)
Similar to lm1 but applied on (st^(1/n)).mv with n=4 

```python
lm1x, rmse1x, mae1x, i_out1x = run_linear_regression_parachor(df_fe0,
                        set(frags)-set(['C#C', 'R3', 'R4', 'R7', 'R8']), 
                        ['C#C', 'R3', 'R4', 'R7', 'R8'],
                        'measured_st',
                        'mv',
                        n=4,
                        run_on_st=False,
                        fit_intercept=True,
                        num_outliers=5,
                        title='fe0 lm1x - surface tension parachor\n(train: liquid, pred: liquid)',
                        x_label='predicted st^(1/4) * mv',
                        y_label='measured st^(1/4) * mv',
                        with_line=True)
print(f"rmse1x = {rmse1x}, mae1x = {mae1x}")
df_fe0.loc[i_out1x]
```

### Linear model - mvpolym

```python
features_exclude_from_mv = features_exclude + ['R5', 'R6']
df_mvpolym_fe0 = df_mvpolym_fe0[df_mvpolym_fe0[features_exclude_from_mv].sum(axis=1) == 0]

X_mvpolym = df_mvpolym_fe0.drop(['M', 'type', 'polymer', 'measured_mv', 'measured_mv_scaled', 'ru']+features_exclude_from_mv, axis=1)
y_mvpolym = df_mvpolym_fe0['measured_mv_scaled']
lm_mvpolym = LinearRegression()
lm_mvpolym.fit(X_mvpolym, y_mvpolym)

y_hat_mvpolym = pd.Series(lm_mvpolym.predict(X_mvpolym), index=X_mvpolym.index)
regression_results(y_mvpolym, y_hat_mvpolym)
```
